###What is Generalization Error and what is its importance in measuring the performance of a machine learning algorithm? Describe at least one model prediction error function each when the supervised learning problem is a regression problem, and an error function when it is a classification problem.

One Error function for regression would be using all the data to fit a continuous output with n datapoints and one continous input where n-1 degrees of freedom to explain the dataset.  The model will fit the each data point perfectly, but if you generalized the model to apply to any further values of inputs you could be seeing a very large source of error.  The reason for this, is you have overfit the data.  For classification, you see the same situation.  Suppose you use a decision tree, but you make sure all the nodes explain every categorical output exactly.  If you were to generalize this decision tree to all future outputs, the decision tree could potentially do a horrible job at predicting the correct category the output should fall into.  

###Explain the bias-variance tradeoff. Describe preference and restriction bias in each of the these algorithms.

The bias variance tradeoff is one of overfitting or underfitting the model.  A model that fits the training data very well, may very well bias the results on the test dataset due to trying to fit the outliers.  You could also be on the other side where, you don't try to overfit overfit the training data, but this causes variance in the test dataset, as it doesn't respond to the valid fluctuations in the true model very well.  

1) SVM: 

The preference bias for SVM would be to allow an objective function that would prefer a simple hypothesis, such as searching all values of gamma, C, and kernel, but penalizing the more extreme results.

Restriction bias would be to only allow a range of possible values for gamma, C and the kernel for SVM.

2) Nerual Networks:

Preference bias would prefer a less complicated model of the activation layers, and nodes.

Restriction bias would only allow permissable range of hidden neuron and activation layers.

3) Decision Trees:

Preference bias would be to prefer smaller decision trees.

Restriction bias would be to limit the number of nodes in the decision tree.

4) Boosting:

Same as Decision Trees

5) k-Nearest Neighbors

Preference bias would prefer smaller values of k, but the penalty function would allow you to search all values for k.

Reference bias would restrist values of k.

###What is cross-validation and when do you use it? What are the other forms of validation techniques?

Cross Validation allows you to ensure that your model can be generalized and perform well in predicting new data that it encounters.  Whenever you are building a model for prediction, cross-validation is appropriate to see if your predictions perform well with the test(s) dataset(s).

Other forms of validation, would be to look at R-Sq adjusted, R-Squared Pred, AIC.  There are cons to all different types of validation methods.  

It seems that using k-folds or holdout methods will provide the best methods for prediction across many hypotheses.

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import StratifiedKFold
import numpy as np
from IPython.display import Image
from sklearn.learning_curve import validation_curve, learning_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.ensemble import GradientBoostingClassifier
from pybrain.utilities           import percentError
from pybrain.tools.shortcuts     import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.structure.modules   import SoftmaxLayer, TanhLayer
from pybrain.datasets            import ClassificationDataSet
from pybrain.structure.modules import Module,SigmoidLayer

def plot(cm, title):
    totals=cm.sum(axis=0)
    correct=cm.diagonal()
    correctPercent=float(correct.sum())/float(totals.sum())
    plt.matshow(cm)
    plt.title(title + "\n")
    plt.text(0,3,"Percent classified correctly: " + str(int(correctPercent*100)) + "%")
    plt.colorbar()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.savefig('confmatrix')
    plt.close()
    return Image(filename='confmatrix.png')
```

# Import Data

```python
headers=['Wine','Alcohol','Malic_Acid','Ash','Alcalinity_Ash', 'Magnesium','Total_Phenols','Flavanoids' \
         ,'Nonflavanoid_Phenols','proanthocyanins','Color_Intesity','Hue','OD280/OD315_DilutedWines','Proline']
```

```python
data=pd.read_csv('wine.csv',header=0,names=headers)
```

```python
X=data.iloc[:,1:14]
y=data.Wine
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25)
```

# SVM  with RBF Kernel

```python
param_range = np.logspace(-6, -1, 5)
train_scores, test_scores = validation_curve(
    SVC(C=1000), X, y, param_name="gamma", param_range=param_range,
    cv=10, scoring="accuracy", n_jobs=1)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.title("Validation Curve with SVM (Kernel=RBF)")
plt.xlabel("$\gamma$")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
plt.semilogx(param_range, train_scores_mean, label="Training score", color="r")
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2, color="r")
plt.semilogx(param_range, test_scores_mean, label="Cross-validation score",
             color="g")
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2, color="g")
plt.legend(loc="best")
#plt.show()
plt.savefig('valCurveSVM')
plt.close()
Image(filename='valCurveSVM.png')
```

```python
train_sizes, train_scores, test_scores = learning_curve(SVC(C=1000,gamma=.00001), X, y, train_sizes=[range(50,150,10)], cv=5)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.title("Learning Curve with SVM (Kernel=RBF)")
plt.xlabel("Training Examples")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
plt.plot(train_sizes, train_scores_mean, label="Training score", color="r")
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2, color="r")
plt.plot(train_sizes, test_scores_mean, label="Cross-validation score",
             color="g")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2, color="g")
plt.legend(loc="best")
#plt.show()
plt.savefig('LC')
plt.close()
Image(filename='LC.png')
```

```python
classifier = SVC(gamma=.00001, C = 1000, degree=3)
y_pred = classifier.fit(X_train, y_train).predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plot(cm,'Confusion Matrix SVM (Kernel=RBF')
```
